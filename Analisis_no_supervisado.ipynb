```python
import numpy as np

from sklearn.cluster import KMeans, DBSCAN
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris
```

```python
X = iris['data']
```

```python
kmeans = KMeans(n_clusters = 3)
predsKmeans = kmeans.fit_predict(X)
predsKmeans
```

```python
kmeans.cluster_centers_
```

```python
dbscan = DBSCAN(eps = 0.6, min_samples = 3)
preds = dbscan.fit_predict(X)
preds
```

PCA para reduccion dimensionalidad

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
X_reducido = pca.fit_transform(X)
X_reducido
```

###Exercise
Con X-reducido, generar un grafico de puntos (scatter) con diferentes colores segun la asignacion provista por KMeans

```python
def graficar(X_reducido):
    import matplotlib.pyplot as plt

    clases = np.unique(preds)

    for clase in clases:
        #Seleccionar los puntos de X_reducido que pertenecen a cada clase
        #Generar el grafico a traves de plt.scatter
        X_clase = X_reducido[preds == clase]
        plt.scatter(X_clase[:,0],X_clase[:,1])
graficar(X_reducido)
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2)
X_reducido_T = tsne.fit_transform(X)
```

```python
graficar(X_reducido_T)
```
