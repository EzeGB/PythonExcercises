Importamos librerias

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
```

Cargamos los datos y sacamos un set para pruebas y otro para entrenar al modelo

```python
iris = load_iris()
X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

Creamos los modelos y los entrenamos

Regresion logistica

```python
logistica = LogisticRegression(max_iter=1000)
log_params = {
    'C': [0.1, 1, 10, 100],
    'solver': ['lbfgs', 'liblinear']
}

gridlog = GridSearchCV(logistica, log_params, cv=5, scoring='accuracy')
gridlog.fit(X_train, y_train)
randomlog = RandomizedSearchCV(logistica, log_params, cv=5, n_iter=10, scoring='accuracy', random_state=42)
randomlog.fit(X_train, y_train)
```

Random Forest

```python
random_Forest = RandomForestClassifier()
forest_params = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}
gridforest = GridSearchCV(random_Forest, forest_params, cv=5, scoring='accuracy')
gridforest.fit(X_train, y_train)
randomforest = RandomizedSearchCV(random_Forest, forest_params, cv=5, n_iter=10, scoring='accuracy', random_state=42)
randomforest.fit(X_train, y_train)
```

Gradient Boosting

```python
gradiente = GradientBoostingClassifier()
paramsGradiente = {
    'n_estimators': [50, 100],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5]
}
gridGradiente = GridSearchCV(gradiente, paramsGradiente, cv=5, scoring='accuracy')
gridGradiente .fit(X_train, y_train)
randomGradiente  = RandomizedSearchCV(gradiente, paramsGradiente, cv=5, n_iter=10, scoring='accuracy', random_state=42)
randomGradiente .fit(X_train, y_train)
```

Comparamos los resultados

```python
print('Precision grid:')
print('grid regresion logistica:',gridlog.best_score_)
print('grid random forest:',gridforest.best_score_)
print('grid gradiente boost:',gridGradiente.best_score_)
```

```python
print('Precision random:')
print('Random regresion logistica:',randomlog.best_score_)
print('Random del random forest:',randomforest.best_score_)
print('Random gradiente boost:',randomGradiente.best_score_)
```

En ambos casos la regresion logistica presenta mejores resultados.
