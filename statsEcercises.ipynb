```python
data = np.array([20, 35, 35, 45, 47, 55, 61, 70, 70, 85])
data = np.sort(data)
```

```python
np.mean(data)
```

```python
np.median(data)
```

```python
stats.mode(data)
```

#exercise
* using a dictionary, count the distribution data
* fin the mode

```python
modeDict = {}
for value in data:
    entries = modeDict.keys()
    if value in entries:
        modeDict[value] = modeDict[value] + 1
    else:
        modeDict[value] = 1
    #OR modeDict[value] = modeDict.get(value, 0) + 1 OMG!!!

print(modeDict)
maxFromModes = max(modeDict.values())
modes = [x for x, count in modeDict.items() if count == maxFromModes]
print(modes)
#turn it into function if you have time
```
