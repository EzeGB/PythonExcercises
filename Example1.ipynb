Match-case Block

```python
day = 10

match day:
    case 1:
        print('Sunday')
    case 2:
        print('Monday')
    case 3:
        print('Tuesday')
    case 4:
        print('Wednesday')
    case 5:
        print('Thursday')
    case 6:
        print('Friday')
    case 7:
        print('Saturday')
    case _:
        print('ERROR: Value out of range.')
```

##Example
* Create a 'days' list which contains all week days.
* Show on screen those list values.

```python
days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

for day in days:
    print('Day',days.index(day)+1, 'nth of the week:',day)
################################################################################
#enumerate usage
print('#############')
daysTwo = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

for index,dayTwo in enumerate(daysTwo):
    print('Index',index, 'Day of the week:',dayTwo)
print('#############')
```

```python
#zip usage

dias = ['Lunes','Martes','Mércoles','Jueves','Viernes','Sábado','Domingo']

for day,dia in zip(days,dias): ##Dependiente de la lista mas corta!
    print(day,":",dia)
```

### Exercise
Create a dictionary with the following elements:
* Hundred: 100
* Two hundred: 200
* ...
* Up to 500

Iterate each key with it's respective value using the items() feature.


```python
numbersDict = {
    'One Hundred':100,
    'Two Hundred':200,
    'Three Hundred':300,
    'Four Hundred':400,
    'Five Hundred':500
}

for text,number in numbersDict.items():
    print(text,"is equivalent to",number)
```

```python
numbersJPDict = {
    '一':1,
    '十':10,
    '百':100,
    '千':1000,
    '万':10000
}

for text,number in numbersJPDict.items():
    print(text,"is equivalent to",number)
```
