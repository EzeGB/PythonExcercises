## Exercise
\begin{align*}
r=\frac{x^2+y^2}{x-y}
\end{align*}

```python
x = 7
y = 5

r = (x ** 2 + y ** 2) / (x - y)
print ('r value:', r)
```

```python
x = 7
y = 5

r = (x * x + y * y) / (x - y)
print ('r value:', r)
#BOTH ways are valid!
```

## Exercise

Define a variable: salary.

* If salary is at least 9000 -> Category A
* If salary is at least 7500 -> Category B
* If salary is at least 5000 -> Category C
* If salary is at least 4500 -> Category D
* Contrary, Category E.

```python
salary = 8500

if salary >= 9000:
  print('Category A')
elif salary >= 7500:
  print('Category B')
elif salary >= 5000:
  print('Category C')
elif salary >= 4500:
  print('Category D')
else:
  print('Category E')
```
