### Distribución exponencial



```python
import numpy as np

from scipy.stats import expon
```

```python
exp = expon(0.5)
exp.cdf(2)
```

```python
exp.cdf(3)
```

```python
exp.cdf(3) - exp.cdf(2)
```

### Ejercicio

El tiempo de atención de un surtidor es en promedio de 5 minutos (dist exponencial).

1. Calcular la probabilidad de que el tiempo sea > 3
2. Probabilidad de que el tiempo esté entre 2 y 4 mins
3. Generar un gráfico de la distribución para este caso

image.png

```python
lam = 1 / 5
x = np.arange(0, 20.1, 0.1)
y = lam * np.exp(-lam * x)
print(x[:10], y[:10])
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y)
```

```python
exp = expon(lam)

# Prob tiempo > 3
1 - exp.cdf(3)
```

```python
exp.cdf(4) - exp.cdf(2)
```
