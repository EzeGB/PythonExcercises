# Requests

Este ejemplo muestra cómo consumir y enviar datos a un endpoint que intercambia datos en formato json

* requests.get
* requests.post
* requests.put
* requests.delete

```python
import requests
import json
import time

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/60.0.3112.113 Safari/537.36'}
result = requests.get(f"https://hubia.bo/applications/apps/estudiantes.php?_={int(time.time())}", headers=headers)
print(result.text)
```

```python
docentes = json.loads(result.text)
docentes
```
