```python
S = [1, 1, 1, 3, 3, 4]
```

##Exercise
1. Create a dict which contains each value from S with its freq.
2. Calculate the probability of 1, 3 and 4.
3. Calculate the expected value.
4. Calculate variance.

```python
dictS = {}
for value in S:
    dictS[value] = dictS.get(value, 0) + 1
dictS
```

```python
dictP = {}
for prob in dictS:
    dictP[prob] = dictS.get(prob)/len(S)
dictP
## Alternativa 0.0
##n = len(S)
## dictS {x: valor / n for x, valor in dictS.items()}
```

```python
expectedValue = 0
for expected in dictP:
    expectedValue += expected * dictP.get(expected)
expectedValue
##Alternativa
##E_X = np.sum([x * p_x for x, p_x in dictS.items()])
```

```python
exSquare = 0
for expected in dictP:
    exSquare += (expected ** 2) * dictP.get(expected)
exSquare
```

```python
variance = exSquare - expectedValue ** 2
variance
```

### Exercise
In a class there are 6 students with approval grades, and 4 failed.
Calculate approval rate.

```python
SClass = ['A', 'A', 'A', 'A', 'A', 'A', 'F', 'F', 'F', 'F']
#A -> 1
#F -> 0
#Reccon p
#Using function pmf, reccon the probabilities of 0 and 1
```

```python
probs = {}
for x in SClass:
    if x == 'A':
        probs[1] = probs.get(1, 0) + 1
    else:
        probs[0] = probs.get(0, 0) + 1
probs
#Alt
#mapping = {'A':1,'R'=0}
#valores = [mapping[valor] for valor in SClass]
```

```python
#Just run once after the previous one!
n = len(SClass)
probs = {x : prob/n for x, prob in probs.items()}
probs
```

```python

newBer = bernoulli(probs.get(1))
print('Prob of A:',newBer.pmf(1))
print('Prob of F:',newBer.pmf(0))
```

```python
newBer.mean()
```

```python
newBer.var()
```

###Binomial Distribution

```python
from scipy.stats import binom
```
