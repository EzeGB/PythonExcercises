1.Dadas dos listas ordenadas, crear un programa que una las dos listas preservando el orden. Ejemplo:

Si lista1 = [1, 3, 4, 5, 7]
Y lista2 = [2, 4, 6, 8]

El resultado debe ser [1, 2, 3, 4, 4, 5, 6, 7, 8]


```python
def tragaListas(listaA,listaB,listaR):
    ## SECCION I: Se compara el valor inicial de las listas, el menor se
    ## añade a la lista Resultado. El primer valor añadido es eliminado
    ## de la lista correspondiente.

    ## SECCION II: Si alguna de las listas ya fue vaciada, todos los elementos
    ## restantes van a la lista resultado vaciando la lista restante.

    ## Si las dos listas estan vacias termina la recursion
    if len(listaA) == 0 and len(listaB) == 0:
        return listaR
    else:
        ##--------INICIO SECCION II--------##
        if len(listaB) == 0:
            for i, elemento in enumerate(listaA):
                listaR.append(listaA[i])
            listaA = []
        elif len(listaA) == 0:
            for i, elemento in enumerate(listaB):
                listaR.append(listaB[i])
            listaB = []
        ##----------FIN SECCION II---------##
        else:
            ##--------INICIO SECCION I--------##
            if listaA[0]<=listaB[0]:
                listaR.append(listaA[0])
                del listaA[0]
            else:
                listaR.append(listaB[0])
                del listaB[0]
            ##----------FIN SECCION I---------##
        ## Al vaciar el contenido de alguna lista se llama a la funcion con
        ## las listas modificadas.
        return tragaListas(listaA,listaB,listaR)

## Un wrapper para que se cumpla el requisito de 2 listas.
def tlWrapper(listaA,listaB):
    listaR = []
    return tragaListas(listaA,listaB,listaR)
```

```python
print('DOS LISTAS VACIAS')
lista1 = []
lista2 = []
print(tlWrapper(lista1,lista2))

print('PRIMERA LISTA VACIA')
lista1 = []
lista2 = [2,4,5]
print(tlWrapper(lista1,lista2))
print('PRIMERA LISTA VACIA, SEGUNDA REPITE ELEMENTOS')
lista1 = []
lista2 = [2,4,5,6,6,6,7]
print(tlWrapper(lista1,lista2))

print('SEGUNDA LISTA VACIA')
lista1 = [2,4,5]
lista2 = []
print('SEGUNDA LISTA VACIA, PRIMERA REPITE ELEMENTOS')
print(tlWrapper(lista1,lista2))
lista1 = [2,4,5,6,6,6,7]
lista2 = []
print(tlWrapper(lista1,lista2))

print('EJEMPLO BASE, UN ELEMENTO REPETIDA EN AMBAS')
lista1 = [1, 3, 4, 5, 7]
lista2 = [2, 4, 6, 8]
print(tlWrapper(lista1,lista2))

print('UNA LISTA MAS GRANDE QUE OTRA POR MAS DE UN ELEMENTO')
lista1 = [1, 3, 4, 5, 7, 8, 9, 10, 50, 100]
lista2 = [2, 4, 6, 8]
print(tlWrapper(lista1,lista2))

print('UNA LISTA MAS GRANDE QUE OTRA POR MAS DE UN ELEMENTO CON REPETIDOS')
lista1 = [1, 2, 2, 3, 4, 5, 7]
lista2 = [2, 4, 6, 8, 8, 8, 10, 12, 5000]
print(tlWrapper(lista1,lista2))

```

2.Crear una función en Python que reciba como parámetro una lista y devuelva una lista con la media de cada k elementos de la misma. Ejemplo:

[1, 2, 3, 4, 5, 6, 7, 8]


Si k = 3, entonces se deberá devolver:

[2, 5, 7.5]

Ya que el promedio de 1, 2, y 3 es 2. El promedio de 4, 5 y 6 es 5, y el promedio de 7 y 8 es 7.5

Si k = 2, entonces se deberá devolver

[1.5, 3.5, 5.5, 7.5]

Ya que el promedio de 1 y 2 es 1.5. El promedio de 3 y 4 es 3.5. El promedio de 5 y 6 es 5.5 y el promedio de 7 y 8 es 7.5.


```python
def promedioSubListas(listaBase,k):
    auxiliar = []
    listaDeListas = []
    promedios = []

    ## Validacion de seguridad
    if k < 1:
        k = 1
    ##Se verifica que la lista no este vacia
    if len(listaBase) != 0:
        ##Se separa la lista en pedazos de tamaño k
        for indice,elemento in enumerate(listaBase):
            if indice % k == 0:
                ##La primera vez no se introduce una lista vacia
                if len(auxiliar) != 0:
                    listaDeListas.append(auxiliar)
                auxiliar = []
            auxiliar.append(elemento)
        ##Se añade la ultima lista restante de tamaño <= k
        listaDeListas.append(auxiliar)
        ##Se itera para sacar el promedio de cada una de las listas
        for lista in listaDeListas:
            suma = 0
            for numero in lista:
                suma += numero
            promedios.append((suma/len(lista)))
    return promedios
```

```python
print('LISTA VACIA')
lista1 = []
k = 3
print(promedioSubListas(lista1,k))

print('LISTA CON NUMEROS IGUALES')
lista1 = [10, 10, 10, 10, 10]
k = 4
print(promedioSubListas(lista1,k))

print('LISTA EJEMPLO 1')
lista1 = [1, 2, 3, 4, 5, 6, 7, 8]
k = 3
print(promedioSubListas(lista1,k))

print('LISTA EJEMPLO 2')
lista1 = [1, 2, 3, 4, 5, 6, 7, 8]
k = 2
print(promedioSubListas(lista1,k))

print('K = 5 EN LISTA ORIGINAL VOLTEADA')
lista1 = [8, 7, 6, 5, 4, 3, 2, 1]
k = 5
print(promedioSubListas(lista1,k))

print('K = 0')
lista1 = [5, 6, 7, 8]
k = 0
print(promedioSubListas(lista1,k))

print('K = 1')
lista1 = [5, 6, 7, 8]
k = 1
print(promedioSubListas(lista1,k))
```
