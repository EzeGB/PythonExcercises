## Clases y objetos

Respecting attributes:

__ (double underscore): Private

_ (single underscore): Protected

no underscore: Public

```python
class Pen:
    def __init__(self,color,thickness,manufacturer):
        self.__color = color
        self.__thickness = thickness
        self.__manufacturer = manufacturer

    def write(self):
        print('Writing...')

    def erase(self):
        print('Erasing...')

    def draw(self):
        print('Drawing...')

    def showAttributes(self):
        print('Color: ' + self.__color +
              ', Thickness: ' + self.__thickness +
              ', Manufacturer: ' + self.__manufacturer)

    @property
    def color(self): #Read: get
        return self.__color

    @property
    def thickness(self):
        return self.__thickness

    @property
    def manufacturer(self):
        return self.__manufacturer

    @color.setter   #Write: set
    def color(self,newColor):
        self.__color = newColor

    @thickness.setter
    def thickness(self,newThickness):
        self.__thickness = newThickness

    @manufacturer.setter
    def color(self,newManufacturer):
        self.__manufacturer = newManufacturer
```

```python
pen1 = Pen('Black','HB','Faber Castell')

pen1.showAttributes()
pen1.erase()

print(pen1.manufacturer)

print(pen1.color)
pen1.color = 'Red'
print(pen1.color)
```

```python
class Person:
    def __init__(self, id, name, lastName):
        self._id = id
        self._name = name
        self._lastName = lastName

    @property
    def id(self):
        return self._id

    @id.setter
    def id(self,newId):
        self._id = newId

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, new_name):
        self._name = new_name

    @property
    def lastName(self):
        return self._lastName

    @lastName.setter
    def lastName(self,new_lastName):
        self._lastName = new_lastName

    def talk(self):
        print('Hello World!')

    def walk(self):
        print('Walking...')

    def show_Personal_Data(self):
        print(f'Id: {self._id}, Name: {self._name}, Last Name: {self._lastName}')
```

```python
class Student(Person):
    def __init__(self, id, name, lastName, level):
        super().__init__(id, name, lastName)  ##Call the father constructor
        self.__level = level

    @property
    def level(self):
        return self.__level

    @level.setter
    def level(self,new_level):
        self.__level = new_level

    def show_Personal_Data(self):
        super().show_Personal_Data()
        print('#######')
        print(f'Level: {self.__level}')
```

```python
person1 = Person('012','John','Smith')
person1.id = '013'

student1 = Student ('11223','Robert','Gomez','First Semester')
student1.id = '22113'

person1.talk()
person1.show_Personal_Data()

student1.walk
print('Student1:',student1.id)
student1.show_Personal_Data()
```

```python
student2 = Student('1111','Walter','White','Second Semester')
student3 = Student('333','Willy','Wonka','Third Semester')
```

```python
students = []
students.append(student1)
students.append(student2)
students.append(student3)
```

```python
for student in students:
    student.show_Personal_Data()
```

### PLAIN TEXTS
open() is used to open files: reading or writing

```python
with open('example.txt','w') as file:
    file.write('Juan,Perez\n') #Writes the name and then a line feed
    file.write('Roberto,Lopez\n')
    file.write('Marta,Gomez')
```

```python
with open('example.txt', 'r') as file:
    for line in file:
        print(line,'#######')
```
