## Exercise
Create a function called and used as a factorial to any number.
0!=1!=1
...
3!=1 * 2 * 3 = 6

```python
def factorial(n:int) -> int: ##You can also strong type it
    if n==0 or n==1:
        return 1
    else:
        return n * factorial(n-1)
```

```python
print(factorial(0))
print(factorial(1))
print(factorial(2))
print(factorial(3))
print(factorial(4))
```

```python
def nonRecursiveFactorial(n):
    result = 1
    if n==0 or n==1:
        return result
    else:
        for value in range (2,n+1):
            result *= value
        return result
```

```python
print(nonRecursiveFactorial(0))
print(nonRecursiveFactorial(1))
print(nonRecursiveFactorial(2))
print(nonRecursiveFactorial(3))
print(nonRecursiveFactorial(4))
```
