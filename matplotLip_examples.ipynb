## Excercise
Knowing that:
* np.exp(x) evaluate exponential function of x
* np.log(x) evaluate natural x logarithm
* np.sin(x) evaluate sin of x
Generate the graphic view of each of these functions generating first all valid vlaues for x

```python
plt.figure(figsize=(5, 3))
plt.title('Gráfico funciones')
plt.grid(True)

x = np.arange(0.1,10+0.1,0.1)
plt.plot(x,np.exp(x)/40, label='exp')
plt.plot(x,100*np.log(x), label='log')
plt.plot(x,100*np.sin(x), label='sin')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()

```

## Exercise

Generate a graphic of sigmoidal function:

\begin{align*}
y=\frac{1}{1+\exp(-x)}
\end{align*}

```python
x = np.arange(-5,5+0.1,0.1)
plt.grid(True)
plt.xticks(np.arange(-5, 5+1, 1))
plt.yticks(np.arange(0, 1+0.1, 0.1))
plt.plot(x,(1/(1+(np.exp(x * -1)))))
```

Gráficos de dispersión

```python
x = np.arange(-5,5+0.1,0.25)
plt.grid(True)
plt.xticks(np.arange(-5, 5+1, 1))
plt.yticks(np.arange(0, 1+0.1, 0.1))
plt.scatter(x,(1/(1+(np.exp(x * -1)))),c='orange',s=5, label='Función Sigmoidal')
plt.legend()
```

### Exercise
Visualize dispersion graphics for the following functions:
* Hyperbolic tangent
* Cosine
* x^2/2

```python
plt.title('Funciones varias')
x = np.arange(-5,5+0.1,0.1)
plt.scatter(x,((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))),c='orange',s=3,label='tanh') #Np tiene tanh ejm
plt.scatter(x,np.cos(x),c='teal',s=5,label='cos')
plt.scatter(x,((x*x)/2)/10,c='orchid',s=5,label='x^2/2')
plt.legend()
```
